# Objective for Part 2

In this part, we will perform EDA on cleaned data to analyse which keywords are most common for misogynistic and non-misogynistic. This will be done using WordCloud.


```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud 
```

```python
# Step 2: Read the cleaned CSV from Part I
df = pd.read_csv("ManualTag_Misogyny_Clean.csv", encoding="latin-1")
```

```python
# Step 3: Combine your values into a long string
long_string = ' '.join(df["cleaned_definition"])
```

```python
long_string
```

```python
# Step 4a: Create a WordCloud object with a 'white' 'background_color'
wordcloud = WordCloud(background_color="white")

# Step 4b: Call the .generate method of the WordCloud object with your string
wordcloud.generate(long_string)

# Step 4c: Use plt to visualize the generated wordcloud
plt.imshow(wordcloud)
plt.show()
```

```python
# Step 5a: Create a WordCloud object with a 'white' 'background_color' and 'max_words' of 50
wordcloud2 = WordCloud(background_color="white", max_words=50)

# Step 5b: Call the .generate method of the WordCloud object with your string
wordcloud2.generate(long_string)

# Step 5c: Use plt to visualize the generated wordcloud
plt.imshow(wordcloud2)
plt.show()
```

```python
# Step 6: Declare two variables, and assign them with DataFrame with either non-misogynist (0) or misogynist text (1)
df_misogyny = df[df['is_misogyny'] == 1]
df_not_misogyny = df[df['is_misogyny'] == 0]
```

```python
# Step 7a: Get combined string from one DataFrame
str_misogyny = ' '.join(df_misogyny["cleaned_definition"])

# Step 7b: Get combined string from the other DataFrame
str_not_misogyny = ' '.join(df_not_misogyny["cleaned_definition"])
```

```python
print(df['is_misogyny'].value_counts())
print(df_misogyny.info())
print(df_not_misogyny.info())
```

```python
# Step 8: Generate word cloud for normal text
wc_normal = WordCloud(background_color="white",max_words=50)

wc_normal.generate(str_not_misogyny)

plt.imshow(wc_normal)
plt.show()
```

### Step 9: Generate wordcloud for misogynist text

```python
# Step 9: Generate word cloud for misogynst text
wc_misogyny = WordCloud(background_color="white",max_words=50)

wc_misogyny.generate(str_misogyny)

plt.imshow(wc_misogyny)
plt.show()
```
